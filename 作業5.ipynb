```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_equation(a, b, c):
    x = np.linspace(-10, 10, 400)  #  x 
    y = a * x**2 + b * x + c  #  y 
    return x, y

# a=-3, b=4, c=7
x1, y1 = quadratic_equation(-3, 4, 7)
# a=5, b=-4, c=9
x2, y2 = quadratic_equation(5, -4, 9)

# 
plt.figure(figsize=(8, 6))

# 
plt.plot(x1, y1, label="y = -3x^2 + 4x + 7")
plt.scatter([-1, 1], [10, 8], color='red', label="Roots: (-1, 10), (1, 8)")
plt.scatter(0.67, 8.11, color='green', label="Vertex: (0.67, 8.11)")

# 
plt.plot(x2, y2, label="y = 5x^2 - 4x + 9")
plt.scatter([0.4, 0.6], [10.4, 9.4], color='red', label="Roots: (0.4, 10.4), (0.6, 9.4)")
plt.scatter(0.5, 9.65, color='green', label="Vertex: (0.5, 9.65)")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Quadratic Equation")
plt.legend()
plt.grid(True)
plt.show()
```
