```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import openpyxl

data = pd.read_csv('population.csv')
data.to_excel('population.xlsx', index=False)

data = pd.read_excel('population.xlsx')
print(data)

plt.plot(data['pop'], data['year'])
plt.xlabel('pop')
plt.ylabel('year')
plt.title('Line Trend Chart')
plt.show()

plt.pie(data['pop'], labels=data['year'], autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
